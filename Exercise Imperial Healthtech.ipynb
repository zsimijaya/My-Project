## Grafik Heatmap

```python
plt.figure(figsize=(10,4),dpi=200)
sns.heatmap(data = df.corr(), cmap = 'hot', annot=True)
```

## Pair Plot

```python
df['Daily User Installs'].mean()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='Date',y='Daily User Installs',data=df,palette='rainbow')
plt.xticks(rotation=90);
plt.title('Jumlah Daily User Installs')
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='Date',y='Daily User Uninstalls',data=df,palette='rainbow')
plt.xticks(rotation=90);
plt.title('Jumlah Daily User Uinstalls')
```

```python
df_mean = pd.DataFrame(df.mean())
```

```python
df_mean
```

```python
df_mean.rename(columns={0:'Nilai Rata-Rata'}, inplace=True)
```

```python
df_mean
```

```python
df_mean.info()
```

```python
df_mean.reset_index(inplace=True)
df_mean = df_mean.rename(columns = {'index':'Kategori'})
```

```python
df_mean
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='Kategori',y='Nilai Rata-Rata',data=df_mean,palette='rainbow')
plt.xticks(rotation=90);
plt.title('Nilai rata-rata Setiap Variabel')
```

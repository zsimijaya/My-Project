```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Melakukan import library yang digunakan untuk pengolahan data
```

# <font color='darkorange'>Membran Sensor 8A, 9A, dan 9B</font>

```python
df = pd.read_csv('HCl2.csv') # Membaca file csv dan memasukkannya kedalam jupyter
```

```python
df = df.dropna() # melakukan seleksi menghilangkan nilai nol (0) pada dataframe
df # menampilkan data bagian awal pada dataframe
```

```python
# Melakukan perhitungan untuk mencari nilai Mean dan Std untuk setiap membran pada Dataframe 

std1 = df.std(numeric_only=True) # menghitung nilai standar deviasi
mean = df.mean(numeric_only=True) #menghitung nilai mean
perhitungan1 = pd.DataFrame(data = std1,columns= ['std']) # Membuat dataframe untuk bagian std 
perhitungan2 = pd.DataFrame(data = mean, columns = ['Mean']) # Membuat dataframe untuk bagian mean
frames = [perhitungan2, perhitungan1] # dataframe mean dan std digabung untuk menjadi satu dataframe
  
result = pd.concat(frames,axis = 1) # Melakukan perintah concat untuk dapat menggabungkan kedua dataframe 
display(result) # Menampilkan hasil dataframe yang telah menjadi satu
```

```python
# Melakukan pemilihan kolom pada dataframe untuk bisa dimasukkan ke dalam grafik
tanggal = df.values[:,0] # Memilih index 0 untuk dijadikan variabel tanggal pengujian pada grafik
sensor = df.values[:, 1:16] # Memilih index 1 sampai 15 untuk dijadikan variabel Tegangan pada grafik
```

```python
# Sourcecode untuk melakukan penampilan grafik

plt.figure(figsize=(14,7),dpi = 300) # Mengatur ukuran grafik dan resolusi tampilan grafik
plt.plot(tanggal, sensor, marker='o') # Melakukan plot tanggal vs sensor 
plt.title("Grafik Plot") # Memberikan judul pada grafik
plt.xlabel('Tanggal Pengujian') # Memberika label pada sumbu x
plt.ylabel('Tegangan (mV)') # Memberikan label pada sumbu y
plt.legend(["8A HCl -38.517333 +/- 24.901899", "9A HCl -47.330000 +/- 27.260157","9B HCl -29.402000 +/- 1.979535","8A NaCl -51.244000 +/- 4.731346","9A NaCl -74.696000 +/- 15.659775", "9B NaCl -52.203333 +/- 1.455212","8A Sukrosa 9.756667 +/- 2.827687","9A Sukrosa -11.646667 +/- 26.828109","9B Sukrosa 8.901333 +/- 2.134691","8A MSG 114.723333 +/- 12.851517","9A MSG 48.889333 +/- 87.674375","9B MSG 117.710667 +/- 14.783185","8A Quinine 115.978667 +/- 19.621618","9A Quinine 76.284000 +/- 46.074660","9B Quinine 107.326000 +/- 32.148337"], loc=(1.03,0.3), fontsize = 11)
# Legend digunakan untuk menampilkan legend
plt.xticks(rotation=90) # Melakukan rotasi 90 derajat pada sumbu x
plt.show() # Menampilkan grafik
```

```python
df_melted = pd.melt(df, id_vars = ['Tanggal'])
df_melted.head(20)
```

```python
plt.figure(figsize=(18,9), dpi = 100)
sns.boxplot(data = df)
plt.xlabel('Larutan')
plt.ylabel('Tegangan (mV)')
plt.title('Grafik Boxplot')
```
